### Download your Data

##### Accession: GSE96678

```python
import urllib.request
import gzip
import shutil
import os

url = "https://ftp.ncbi.nlm.nih.gov/geo/series/GSE96nnn/GSE96678/matrix/GSE96678_series_matrix.txt.gz"
downloaded_file = "GSE96678.gz"
unzipped_file = "GSE96678.txt"

# Download the gzipped file
urllib.request.urlretrieve(url, downloaded_file)

# Unzip the file
with gzip.open(downloaded_file, 'rb') as f_in:
    with open(unzipped_file, 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)

print(f"File downloaded and unzipped to: {unzipped_file}")

```
